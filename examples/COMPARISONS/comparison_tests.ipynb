We will plot some figure showing how fast we are and how much further we scale.

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
comparison = 'exp'

with open(f'results/comparison_test_{comparison}_device_-1.pkl', 'rb') as f:
    cpu_times = pickle.load(f)

with open(f'results/comparison_test_{comparison}_device_0.pkl', 'rb') as f:
    gpu_times = pickle.load(f)

plt.plot(cpu_times, label='addition_cpu', color='r')
plt.plot(gpu_times, label='addition_gpu', color='b')
plt.xlabel('Bitwidth')
plt.ylabel('Time (s)')
plt.yscale('log')
plt.ylim(1e-3, 1e2)
plt.legend()
plt.tight_layout()
plt.show()
```
